#### Investigate relationships between variables 

We'll use a chi-square test to test the relationships between the features and the label

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import chi2
```

```python
labelencoder = LabelEncoder()
```

```python
# Encode the categorical variables
dataset_encoded = pd.DataFrame()
dataset_encoded['workclass'] = labelencoder.fit_transform(dataset['workclass'])
dataset_encoded['education-num'] = labelencoder.fit_transform(dataset['education-num'])
dataset_encoded['marital-status'] = labelencoder.fit_transform(dataset['marital-status'])
dataset_encoded['occupation'] = labelencoder.fit_transform(dataset['occupation'])
dataset_encoded['relationship'] = labelencoder.fit_transform(dataset['relationship'])
dataset_encoded['race'] = labelencoder.fit_transform(dataset['race'])
dataset_encoded['sex'] = labelencoder.fit_transform(dataset['sex'])
dataset_encoded['native-country'] = labelencoder.fit_transform(dataset['native-country'])
dataset_encoded['label'] = labelencoder.fit_transform(dataset['>50K'])
```

```python
chi_scores = chi2(dataset_encoded.drop('label', axis=1), dataset_encoded['label'])
```

```python
chi_scores
```

```python
p_values = pd.Series(chi_scores[1], index = dataset_encoded.drop('label', axis=1).columns)
p_values.sort_values(ascending=False, inplace=True)
```

```python
p_values.plot.bar()
```

The column workclass has the highest chi-square statistic. This means that it is more likely to be independent from the label (the chi-square test tests for independence between categorical variables). So we can probably drop this feature. 
More information about how chi-squared tests are done: https://www.statisticshowto.com/probability-and-statistics/chi-square/#chi-square%20statistic

Let's plot the output of the number of instances per workclass where the label is >50K and <50K

```python
workclass_analysis = dataset[['workclass','>50K','sex']].groupby(['workclass','>50K']).count() # since we're counting the number of rows per group, we could have taken any column. sex was arbitrarily chosen
workclass_analysis.columns = ['count']
workclass_analysis = workclass_analysis.reset_index() # to change the indexes workclass and >50K to columns so that they can be plotted in seaborn 
workclass_analysis
```

```python
sns.set_theme(style="whitegrid")
sns.barplot(data=workclass_analysis, x='workclass', y='count', hue='>50K')
plt.xticks(rotation=90)
plt.show()
plt.style.use('default')
```

Looks like the number of rows that belong to the workclass 'Private' is high, and therefore it has many instances where the income exceeds 50K, as well as many instances where income is less than 50K (relative to the other workclasses). In other words, the proportion of rows with income greater than 50K and less than 50K seems to be similar between the workclasses (take a look at Private, Local-gov and Self-emp-not-inc), and thus does seem not have any helpful information to classify each row to >50K or <50K. So it seems to make sense that the label has no dependence on the workclass.

#### 1.5.Convert categorical variables from string format to binary (preparation for classification task). Apply one-hot encoding.  

```python
dataset.info()
```

We can see that the columns 'education', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'native-country' and '>50K' have categorical variables.
We'll now one-hot encode these variables concatenate them to the dataframe, and then drop the categorical columns

```python
# Create the one-hot dataframes
# workclass_onehot = pd.get_dummies(dataset['workclass'], prefix='workclass', drop_first=True)
education_onehot = pd.get_dummies(dataset['education'], prefix='education', drop_first=True)
marital_status_onehot = pd.get_dummies(dataset['marital-status'], prefix='marital-status', drop_first=True)
occupation_onehot = pd.get_dummies(dataset['occupation'], prefix='occupation', drop_first=True)
relationship_onehot = pd.get_dummies(dataset['relationship'], prefix='relationship', drop_first=True)
race_onehot = pd.get_dummies(dataset['race'], prefix='race', drop_first=True)
sex_onehot = pd.get_dummies(dataset['sex'], prefix='sex', drop_first=True)
native_country_onehot = pd.get_dummies(dataset['native-country'], prefix='native-country', drop_first=True)
label = pd.get_dummies(dataset['>50K'], drop_first=True, columns=['label'])
```

```python
# Concatenate the one hot columns to create a new dataframe
dataset_onehot = pd.concat((dataset, education_onehot, marital_status_onehot, occupation_onehot, relationship_onehot, race_onehot, sex_onehot, native_country_onehot, label), axis=1)
```

```python
# Drop the categorical columns
dataset_onehot = dataset_onehot.drop(['education', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'native-country', '>50K'], axis=1)
```

### Part 2. Decision tree creation and performance evaluation 

<a class="anchor" id="2.1"></a>
#### 2.1.The selected and prepared dataset should be divided into subsamples for training and validation. Cross-validation is allowed.

As shown above, the data is split into train and test sets. We're not going to use cross-validation. We'll just use the test set to measure the accuracy of the model after training is complete 

<a class="anchor" id="2.2"></a>
#### 2.2.Implement/run an algorithm for building a decision tree (recommended "Iterative Dichotomiser" algorithm)

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier(splitter='best', max_depth=34, max_leaf_nodes=15, min_samples_split=14)
```

Let's use GASearchCV to find the best parameters to pass to the model, and then train the model using the best parameters

```python
from sklearn_genetic import GASearchCV
from sklearn_genetic.space import Categorical, Integer, Continuous
```

```python
param_grid = {'splitter': Categorical(['best','random']),
             'max_depth': Integer(2,50),
             'max_leaf_nodes': Integer(2,75),
             'min_samples_split': Integer(2,15)}
```

```python
evolved_estimator = GASearchCV(estimator=model,
                             scoring='accuracy',
                             param_grid=param_grid,
                             n_jobs=2,
                             verbose=True)
```

Let's add a callback to prevent the model from training for too long without any noticeable improvement

```python
from sklearn_genetic.callbacks import DeltaThreshold
callback = DeltaThreshold(threshold=0.001, generations=5, metric='fitness_min')
```

```python
evolved_estimator.fit(X_train, y_train, callbacks=callback)
```

```python
evolved_estimator.best_params_
```

We'll recreate the model using the best params

```python
model = tree.DecisionTreeClassifier(splitter='best', max_depth=11, max_leaf_nodes=60, min_samples_split=14)
```

<a class="anchor" id="2.3"></a>
#### 2.3.Train the algorithm on the training set and evaluate the performance on the test set. 

```python
model.fit(X_train,y_train)
```

```python
prediction = model.predict(X_test)
```

```python
prediction
```

Accuracy score on the test set:

```python
model.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
confusion_matrix(y_test, prediction)
```

```python
accuracy_score(y_test, prediction) # same as model.score(X_test, y_test)
```

We get an accuracy of 85.52% on the test set
